```python
import pandas as pd
df=pd.read_html('https://www.imdb.com/chart/top')
len(df)
```

```python
df=df[0]
```

```python
df['Rank & Title']=df['Rank & Title'].str.strip(')')
df['Rank']=df['Rank & Title'].str.split('.').str[0]
df['Year']=df['Rank & Title'].str.split('(').str[1]
df['Rank and Title']=df['Rank & Title'].str.split('(').str[0]
df['Title']=df['Rank and Title'].str.split('.').str[1]
df
```

```python
df=df.drop(['Unnamed: 0','Rank & Title','Your Rating','Unnamed: 4'],axis=1)
```

```python
df=df[['Rank','Title','Year','IMDb Rating']]
```

```python
df.to_csv('IMDb Rating.csv',index=False)
```
